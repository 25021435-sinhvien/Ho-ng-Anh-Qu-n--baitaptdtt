W4A10[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất.

```python
def get_collatz_length(num):
    # """Hàm tính độ dài dãy Collatz của một số"""
    count = 1
    while num>1:
        if num%2==0:
            num=num//2
        else:
            num=3*num+1
        count+=1
    return count
def solve():
    try:
        line = input().strip()
        if not line:
            return
        n = int(line)
    except (EOFError, ValueError):
        return
    max_len = 0
    best_x = 0
    for i in range(1,n+1):
        current_len = get_collatz_length(i)
        if current_len>max_len:
            max_len=current_len
            best_x=i
    print(f"{best_x} {max_len}")
if __name__ == "__main__":
    solve()

```

```python
def get_collatz_length(num):
    """Hàm tính độ dài dãy Collatz của một số"""
    count = 1
    while num > 1:
        if num % 2 == 0:
            num = num // 2
        else:
            num = 3 * num + 1
        count += 1
    return count

def solve():
    try:
        # Nhập n từ bàn phím
        line = input().strip()
        if not line:
            return
        n = int(line)
    except (EOFError, ValueError):
        return

    max_len = 0
    best_x = 0

    # Duyệt từ 1 đến n
    for i in range(1, n + 1):
        current_len = get_collatz_length(i)
        
        # Nếu độ dài mới LỚN HƠN hẳn kỷ lục cũ thì mới cập nhật
        # Điều này giúp giữ lại số i nhỏ nhất nếu có trùng độ dài
        if current_len > max_len:
            max_len = current_len
            best_x = i

    # In kết quả theo định dạng: x L
    print(f"{best_x} {max_len}")

# Chú ý: 2 dấu gạch dưới ở mỗi bên __name__ và __main__
if __name__ == "__main__":
    solve()
```

Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
import math
def tong_thuc(n):
    if n<2: return 0
    tong = 1
    # can_n = int(math.sqrt(n))
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            tong+=i
            if n//i!=i:
                tong+=n//i
    return tong
a = int(input())
b = int(input())
if a>0 and b>0 and a!=b and tong_thuc(a)==b and tong_thuc(b)==a:
    print("True")
else:
    print("False")        
```

```python
import math

def tong_uoc_thuc_su(n):
    if n < 2: return 0
    # Số 1 luôn là ước thực sự của mọi số n > 1
    tong = 1 
    can_n = int(math.sqrt(n))
    
    # Duyệt từ 2 đến căn bậc hai của n
    for i in range(2, can_n + 1):
        if n % i == 0:
            tong += i       # Cộng ước nhỏ
            if i != n // i: # Nếu n không phải số chính phương (ví dụ 36 = 6x6)
                tong += n // i # Cộng ước lớn tương ứng
    return tong

# Nhập dữ liệu
try:
    a = int(input("Nhập số thứ nhất: "))
    b = int(input("Nhập số thứ hai: "))

    # Kiểm tra điều kiện cặp số thân thiết (a != b để loại trừ số hoàn hảo)
    if a > 0 and b > 0 and a != b and tong_uoc_thuc_su(a) == b and tong_uoc_thuc_su(b) == a:
        print("true")
    else:
        print("false")
except ValueError:
    print("Vui lòng nhập số nguyên dương.") 
```

W4A14[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
def find_ucln(m, n ):
    while n!=0:
        m, n = n, m%n
    return m
m = int(input())
n = int(input())
print(find_ucln(m, n))
```

```python
def find_gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

# Nhập dữ liệu
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))

# Xuất kết quả
print(find_gcd(m, n))
```
