W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int,input().split())
s = min(a,b)
for i in range(1,s+1):
    if a%i==0 and b%i==0:
        print(i,end=' ')
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n=int(input())
chan=0
le=0
while n>0:
    s=n%10
    if s%2==0:
        chan+=1
    else:
        le+=1
    n//=10
print(chan, le)
```

W4A24 Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a=int(input())
n=0
s=0
while s<=a:
    n+=1
    s+=1/n
print(f'n nhỏ nhất là:{n}')
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
max_d = None
min_d = None
while True:
    n=int(input())
    if n==-1:
        break
    elif max_d is None:
        max_d=n
        min_d=n
    else:
        if n>max_d:
            max_d=n
        if n<min_d:
            min_d=n
if max_d is not None:
    print(f'so lon nhat la:{max_d}')
    print(f'so nho nhat la:{min_d}')
else:
    pass

```

W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
a = int(input())
f1=1
f2=1
while f1+f2<a:
    f1=f2
    f2=f1+f2
print(f2)
```
