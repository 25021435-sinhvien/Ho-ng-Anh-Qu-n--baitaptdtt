Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a = float(input())
b = float(input())
c = float(input())
p = (a+b+c)/2
s = (p*(p-a)*(p-b)*(p-c))**(1/2)
if a+b>c and a+c>b and b+c>a and min(a,b,c)>0:
    if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print(f"là tam giác vuông và diện tích là {s} và chu vi là {2*p}")
    elif a == b or a == c or b == c:
        if a**2 + b**2==c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            print(f"là tam giác vuông cân và diện tích là {s} và chu vi là {2*p}")
        else:
            print(f"là tam giác cân và diện tích là {s} và chu vi là {2*p}")
    elif a == b == c:
        print(f"là tam giác đều và diện tích là {s} và chu vi là {2*p}")
    else:
        print("la tam giác thường và diện tích là {s} và chu vi là {2*p}")
else:
    print("không là tam giác")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
