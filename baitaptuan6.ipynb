W6A9 Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key. 


```python
                                    #  BAI THAM KHAO
# 1. Hàm để nhập dữ liệu cho một dictionary từ bàn phím
def nhap_dict(ten_dict):
    d = {}
    print(f"--- Nhập dữ liệu cho {ten_dict} ---")
    s = input("Nhập các cặp key:value (ví dụ: a:10 b:20): ")
    if s.strip():
        pairs = s.split()
        for p in pairs:
            key, val = p.split(':')
            d[key] = int(val)
    return d

# 2. Nhập hai dictionary
dict1 = nhap_dict("Dictionary 1")
dict2 = nhap_dict("Dictionary 2")

# 3. Gộp hai dictionary
# Tạo một bản sao của dict1 để làm dictionary kết quả
res_dict = dict1.copy()

for key, val in dict2.items():
    if key in res_dict:
        # Nếu key đã tồn tại trong dict1, cộng dồn giá trị
        res_dict[key] += val
    else:
        # Nếu key chưa có, thêm mới vào
        res_dict[key] = val

# 4. Sắp xếp theo thứ tự từ điển của Key
# sorted() sẽ trả về một danh sách các Key đã được sắp xếp
sorted_keys = sorted(res_dict.keys())

# Tạo dictionary mới từ danh sách key đã sắp xếp
final_dict = {}
for k in sorted_keys:
    final_dict[k] = res_dict[k]

# 5. In kết quả
print("\nDictionary sau khi gộp và sắp xếp:")
print(final_dict)
```

W6A10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.


```python
num=[int(x) for x in input().split()]
k=int(input())
chuoi=[]
n=len(num)
for i in range(n):
    for j in range(i+1,n):
        # ta taoj them j de tranh viec lap lai cac so truoc i
        if num[i]+num[j]==k:
            chuoi.append((num[i],num[j]))
chuoi.sort()
# sort se xep cac tuple lan luot tu be den lon o cac so dau tien
print(chuoi)
```

[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.

```python
num=[int(x)for x in input().split()]
if not num:
    pass
else:
    cnt={}
    for i in num:
        cnt[i]=cnt.get(i,0)+1
        # neu gap 1 so lan dau thi cnt.get(i,0) tra ve 0 sau do cong 1 ben canh tu lan 2 tro di moi lan gap se cong them 1
    max_n=max(cnt.values())
    candidates=[]
    for num,b in cnt.items():
        # items se bien dictionary thanh cap tuple vd 3:2 -> (3,2) tuwf ddos num,b se gans lan luot tu trai sang phai
        if b==max_n:
            candidates.append(num)
    print(min(candidates))
```

```python
num=[int(x)for x in input().split()]
if not num:
    pass
else:
    maxl=0
    for i in num:
        s=num.count(i)
        if s>maxl:
            maxl=s
    ds=[]
    for i in num:
        if num.count(i)==maxl:
            ds.append(i)
    print(min(ds))

```

[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban

```python
a=input().split()
old={}
for i in a:
        k,v=i.split(':')
        old[k]=v
new={}
for k,v in old.items():
    new[v]=k
print(new)
```

```python
a=input().split()
old={}
for i in a:
    key,v=i.split(':')
    old[key]=int(v)
k=int(input())
new={}
for key,v in old.items():
    if v>k:
        new[key]=v
print(new)

    
```
