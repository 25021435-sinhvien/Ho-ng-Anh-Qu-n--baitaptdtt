Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
Ví dụ: Ta nhập h = 4:
   *
  * *
 *   *
*******

```python
n = int(input())
for i in range(1,n+1):
    print(" "*(n-i),end="")
    if i==1:
        print("*")
    elif i==n:
        print("*"*(2*n-1))
    else:
        print("*"+" "*(2*i-3)+"*")
```

```python
n = int(input("Nhập h: "))

for i in range(1, n + 1):
    # 1. In khoảng trắng bên ngoài để căn giữa
    print(" " * (n - i), end="")
    
    # 2. Xử lý phần thân tam giác
    if i == 1:
        # Dòng đầu tiên: Chỉ có 1 dấu sao
        print("*")
    elif i == n:
        # Dòng cuối cùng: In đặc toàn bộ (2*n - 1 dấu sao)
        print("*" * (2 * n - 1))
    else:
        # Các dòng ở giữa: In 1 sao + khoảng trống + 1 sao
        # Công thức khoảng trống giữa: 2*i - 3
        print("*" + " " * (2 * i - 3) + "*")
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
import math
a = int(input())
s = int(math.sqrt(a))
if a%s==0:
    print(f"{a} la so chinh phuong")
else:
    print(f"{a} khong la so chinh phuong")
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
max_d = None
min_d = None
while True:
    n = int(input())
    if n==-1:
        break
    if max_d is None:
        max_d=n
        min_d=n
    else:
        if n>max_d:
            max_d=n
        if n<min_d:
            min_d=n
if max_d is not None:
    print(f"số lớn nhất là:{max_d}")
    print(f"số nhỏ nhất là:{min_d}")
else:
    pass

```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
n = int(input())
f1 = 1
f2 = 1
while f1+f2<=n:
    f = f1 + f2
    f1 = f2
    f2 = f
print(f2)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input())
so_nt = [2,3,5,7]
cnt = 0
while n>0:
    s=n%10
    if s in so_nt:
        cnt+=1
    n//=10
print(cnt)
```
