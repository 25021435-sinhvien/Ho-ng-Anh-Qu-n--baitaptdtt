W3A1 Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input())
if 100 <= n <= 999:
    tram = n//100
    chuc = (n%100)//10
    donvi = n%10
    print(f"{donvi}{chuc}{tram}")
else:
    pass
```

W3A11 [Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm
tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều",
"Tam giác cân", hay "Tam giác thường".

```python
a = int(input())
b = int(input())
c = int(input())
if a+b>c and a+c>b and b+c>a and min(a,b,c)>0:
    if a==b==c:
        print("là tam giác đều")
    elif a==b!=c or a==c!=b or c==b!=a:
        print("là tam giác cân")
    else:
        print("là tam giác thường")
else:
    print("không phải tam giác")
```

W3A16 [Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách

```python
n = float(input())
phan_nguyen = int(n)
phan_thap_phan = n - phan_nguyen
if n>0 or phan_thap_phan==0:
    xuong = phan_nguyen
else:
    xuong = phan_nguyen - 1
if n<0 or phan_thap_phan==0:
    len = phan_nguyen
else: 
    len = phan_nguyen + 1
if phan_thap_phan>=0.5:
    gan = phan_nguyen + 1
else: 
    gan = phan_nguyen
print(f"{len} {xuong} {gan}")

```
