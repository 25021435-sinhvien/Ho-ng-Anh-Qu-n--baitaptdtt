Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n=int(input())
m=int(input())
matrix=[]
for i in range(n):
    row=[]
    for j in range(m):
        a=int(input())
        row.append(a)
    matrix.append(row)
for row in matrix:
    for a in row:
        print(f"{a:4d}",end='')
    print()
```

W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n=int(input())
matrix=[]
for i in range(n):
    row=list(map(int,input().split()))
    matrix.append(row)
c=[]
for i in range(n):
    c.append(matrix[i][i])
p=[]
for i in range(n):
    p.append(matrix[i][n-1-i])
print(c)
print(p)
```

W6A18
Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k)

```python
n,m,k=map(int,input().split())
matrix=[]
for i in range(n):
    row=[]
    for j in range(m):
        a=int(input())
        row.append(a)
    matrix.append(row)
if 0<=k<n:
    tongh=sum(matrix[k])
else:
    tongh=0
tongc=0
if 0<=k<m:
    for i in range(n):
        tongc+=matrix[i][k]
else:
    tongc=0
print(tongh,tongc)
```

W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
def timso(num,k):
    for i in num:
        if num[i]==k:
            return i
    return -1
num=[int(x) for x in input().split()]
k=int(input())
print(timso(num,k))
```

W6A23 VIết hàm xóa đi các số <x có trong mảng

```python
def xoanhohon1(ds,x):
    new=[]
    for i in ds:
        if i>x:
            new.append(i)
    return new
ds=[int(i) for i in input().split()]
x=int(input())
print(xoanhohon1(ds,x))
```

```python
def xoanhohon2(num,x):
    new=[int(i) for i in num if i>=x]
    return new
num=[int(i) for i in input().split()]
x=int(input())
print(xoanhohon2(num,x))
```

W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 

```python
def xeplai(num):
    z=[i for i in num if i==0]
    k=[x for x in num if x!=0]
    return z+k
num=list(map(int,input().split()))
print(xeplai(num))
```

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1

```python
def inra(num,x):
    k=[]
    for i in num:
        if i<x:
            k.append(i)
        elif len(k)==0:
            return -1
    return k
      
num=[int(x) for x in input().split()]
x=int(input())
print(inra(num,x))
         
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
num=[int(x) for x in input().split()]
n=len(num)
dgiam=True
dtang=True

for i in range(n-1):
    if num[i]>=num[i+1]:
        dtang=False
    if num[i]<=num[i+1]:
        dgiam=False
    
if dtang:
    print('day tang')
elif  dgiam:
    print('day giam')
else:
    print('day vo danh')
  
```
