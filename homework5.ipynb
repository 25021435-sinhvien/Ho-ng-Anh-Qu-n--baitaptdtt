W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_cnt(a,b,c):
    d=b**2-4*a*c
    if d>=0:
        x1=(-b-d**0.5)/2*a
        x2=(-b+d**0.5)/2*a
        return x1,x2
    else:
        return 'phuong trinh vo nghiem'
a,b,c=map(float,input().split())
ket_qua=quadratic_roots_cnt(a,b,c)
print(ket_qua)
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def arecollinear(x1,y1,x2,y2,x3,y3):
    if (x2-x1)*(y3-y1)==(x3-x1)*(y2-y1):
        return True
    return False
x1,y1,x2,y2,x3,y3=map(float,input().split())
ket_qua=arecollinear(x1,y1,x2,y2,x3,y3)
print(ket_qua)
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def goc_B(x1,y1,x2,y2,x3,y3):
    BA=(x1-x2,y1-y2)
    BC=(x3-x2,y3-y2)
    vohuong=BA[0]*BC[0]+BA[1]*BC[1]
    dodai=math.sqrt(((x1-x2)**2+(y1-y2)**2)*((x3-x2)**2+(y3-y2)**2))
    if (x1-x2)**2+(y1-y2)**2==0 or (x3-x2)**2+(y3-y2)**2==0:
        return 0
    cosb=vohuong/dodai
    cosB=max(-1,min(1,cosb))
    rad=math.acos(cosB)
    deg=math.degrees(rad)
    return deg
x1,y1,x2,y2,x3,y3=map(float,input().split())
ket_qua=goc_B(x1,y1,x2,y2,x3,y3)
print(ket_qua)

```

W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_10_n(n):
    if n<=0:
        return 0
    else:
        cnt=0
        for x in range(1,n+1):
            s=str(x)
            if len(s)==len(set(s)):
                cnt+=x
        return cnt
n=int(input())
ket_qua=sum_1_10_n(n)
print(ket_qua)
        
```
